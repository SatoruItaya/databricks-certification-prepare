```
from pyspark.sql import Row
```

```
import datetime
users = [
    {
        "id": 1,
        "first_name": "Corrie",
        "last_name": "Van den Oord",
        "email": "cvandenoord@etsy.com",
        "phone_numbers": Row(mobile = "+0 000 000 000", home = "+0 111 111 111"),
        "is_customer": True,
        "amount_paid": 1000.55,
        "customer_from": datetime.date(2021,1,15),
        "last_updated_ts": datetime.datetime(2021, 2, 10, 1, 15, 0)
    },
    {
        "id": 2,
        "first_name": "Nikolaus",
        "last_name": "Brewitt",
        "email": "brewitt@etsy.com",
        "phone_numbers": Row(mobile = "+1 000 000 000", home = "+1 111 111 111"),
        "is_customer": True,
        "amount_paid": 900.55,
        "customer_from": datetime.date(2021,2,15),
        "last_updated_ts": datetime.datetime(2021, 2, 17, 1, 15, 0)
    },
    {
        "id": 3,
        "first_name": "Nikolaus",
        "last_name": "Brewitt",
        "email": "brewitt@etsy.com",
        "phone_numbers": Row(mobile = None, home = None),
        "is_customer": True,
        "amount_paid": 900.55,
        "customer_from": datetime.date(2021,2,15),
        "last_updated_ts": datetime.datetime(2021, 2, 17, 1, 15, 0)
    }
]
```

```
users_df = spark.createDataFrame([Row(**user) for user in users])
```

```
users_df.show()
```

```
users_df.select("id", "phone_numbers").show(truncate=False)
```

```
users_df.select("id", "phone_numbers.mobile", "phone_numbers.home").show(truncate=False)
```

```
from pyspark.sql.functions import col
```

```
users_df.select("id", col("phone_numbers")["mobile"], col("phone_numbers")["home"]).show()
```

```
users_df.select("id", "phone_numbers.*").show(truncate=False)
```
