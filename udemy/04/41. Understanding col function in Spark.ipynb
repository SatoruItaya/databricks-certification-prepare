```
from pyspark.sql.functions import date_format
```

```
users_df.select(
    col("id"),
    date_format("customer_from", "yyyyMMdd")
).show()
```

```
users_df.select(
    col("id"),
    date_format("customer_from", "yyyyMMdd")
).printSchema()
```

```
users_df.select(
    col("id"),
    date_format("customer_from", "yyyyMMdd").alias("customer_from")
).show()
```

```
users_df.select(
    col("id"),
    date_format("customer_from", "yyyyMMdd").cast("int").alias("customer_from")
).show()
```

```
users_df.select(
    col("id"),
    date_format("customer_from", "yyyyMMdd").cast("int").alias("customer_from")
).printSchema()
```
